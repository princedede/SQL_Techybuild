```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('filmss.csv', sep=';')
df
```

```python
df['gross'] = df['gross'].astype(str)
df['budget'] = df['budget'].astype(str)
```

```python
df['gross'] = df['gross'].str[:-2]
```

```python
df['budget'] = df['budget'].str[:-2]
```

```python
df
```

```python
df['budget'] = df['budget'].replace('n',0)
```

```python
df['gross'] = df['gross'].replace('n',0)
```

```python
df
```

```python
df['gross'] = df['gross'].astype(int)
```

```python
df['budget'] = df['budget'].astype(float)
```

```python
df.info()
```

```python
df['gross'] = np.where(df['gross'] == 0, np.NaN, df['gross'])
```

```python
df['budget'] = np.where(df['budget'] == 0, np.NaN, df['budget'])
```

```python
df
```

```python
df.info()
```

```python
df.to_csv('films_new.csv', index=False)
```
